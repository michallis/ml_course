## Show gender onboarding depending on the Embarked location

Average age for female and male

```python
bySexAge = df[["Sex","Age","Embarked"]]
grp = bySexAge.groupby(["Embarked","Sex"]).agg({'Age' : 'mean'})
grp
```

Show average age for passengers based on embarked location

```python
byEmbarked = df[["Age","Embarked"]]
grp = byEmbarked.groupby("Embarked",as_index=True).agg({'Age' : 'mean'})
grp
```

```python
x = grp["Age"]
y = grp["Embarked"]
ylabels = ["C","Q","S"]
plt.plot(x,y,"y^")
plt.yticks(range(len(y)), ylabels, size='small')
plt.xlabel('Average Age')
plt.ylabel('Embark Location')
plt.title("Average Age per Embark Location")
plt.show()
```
