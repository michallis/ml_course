Total Years of memoreis lost (using average of passengers where age is known)

```python
totalPassengers = len(df)
agesDf = df[["Age"]].dropna()
passengersAgeKnown = len(df[df["Age"] > 0])
totalKnownYearsLost = sum(agesDf["Age"])
avgYearsLivedForPassengersWithAge = sum(agesDf["Age"])/passengersAgeKnown
averageTotalYearsLost = totalKnownYearsLost + ((totalPassengers - passengersAgeKnown)*avgYearsLivedForPassengersWithAge)
averageTotalYearsLost
```
